```python
from sympy import *
```

```python
g = lambda x: (x + 1)**(1/5)
```

```python
x = symbols('x')
```

```python
plot(g(x), (x, 1, 3))
```

```python
plot(abs(g(x).diff()), (x, 1, 3))
```

```python
q = 0.12
```

```python
k = 0
x0 = 3
while k < 50:
    k += 1
    x = g(x0)
    e = q / (1 - q) * abs(x - x0)
    x0 = x
    if e < 10**-5:
        print('*', k, x, e)
        break
    print(k, x, e)
    
```
